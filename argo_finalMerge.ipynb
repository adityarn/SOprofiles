```python
import pandas as pd
import numpy as np
import glob
```

```python
# Get a list of all CSV files in the directory
path = '/scratch/an1n23/argo_RS_AS/'  # Change to your path
all_files = glob.glob(path + "/*.part")
```

```python
all_files
```

```python
dfs = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    dfs.append(df)

argo = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
argo
```

```python
argo.to_csv("/home/an1n23/datasets/argo_RS_AS_merged.csv", index=False)
```
